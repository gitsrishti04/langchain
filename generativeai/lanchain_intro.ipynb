```python
##streaming and batch processing
```

```python
model.invoke("Write me a 500 words paragraph on AI")
```

```python
for chunk in model.stream("Write a 500 words paragraph on AI"):
    print(chunk.content, end="|", flush = True)
```

```python
responses= model.batch(
    [
        "why do parrots talk?",
        "how do airplanes fly?",
        "what is quantum computing?"
    ]
)
for response in responses:
    print(response)
```

```python
model.batch(
    ["why do parrots talk?",
        "how do airplanes fly?",
        "what is quantum computing?"
    ],
    config={
        'max_concurrency':5 , # limit to 5 parallel calls
    }
)
```
